# NEOFC - Prepare phenotypic data for HCP-Young Adult dataset

```python
from pathlib import Path 
import pandas as pd 
import numpy as np 
from nispace.io import read_json

# working directory
wd = Path.cwd() 
print(wd)
```

## Subjects

# mapFC - Prepare physio data for HCP-Young Adult

```python
subs = np.loadtxt(wd / "data_source" / "pheno" / "hcp_ya" / "hcp_ya_subjects.txt", dtype=int)
print(f"Number of subjects: {len(subs)}")
```

### Get HCP-YA phenotyping data

```python
hcp_ya_pheno = pd.concat([
    pd.read_csv(wd / "data_source" / "pheno" / "hcp_ya" / "hcp_ya_pheno_open.csv", index_col=0),
    pd.read_csv(wd / "data_source" / "pheno" / "hcp_ya" / "hcp_ya_pheno_restricted.csv", index_col=0)
], axis=1)

# subjects
hcp_ya_pheno = hcp_ya_pheno.loc[subs]

# basic variables to retain
hcp_ya_pheno = hcp_ya_pheno[["Gender", "Age_in_Yrs", "Family_ID", "BMI", "CogFluidComp_Unadj", "CogCrystalComp_Unadj", "CogTotalComp_Unadj"]] 

# rename
hcp_ya_pheno.index.name = "sub"
hcp_ya_pheno = hcp_ya_pheno.rename(columns={
    "Gender": "sex", 
    "Age_in_Yrs": "age", 
    "Family_ID": "family_id", 
    "BMI": "bmi",
    "CogFluidComp_Unadj": "cog_fluid",
    "CogTotalComp_Unadj": "cog_total",
    "CogCrystalComp_Unadj": "cog_crystal"
})

# make sure all subjects unrelated
print(f"{hcp_ya_pheno['family_id'].nunique()}/{hcp_ya_pheno.shape[0]} subjects are unrelated")

# Print gender ratio and age distribution
print(hcp_ya_pheno["sex"].value_counts(normalize=True))
print(hcp_ya_pheno["age"].describe())

# save
hcp_ya_pheno.to_csv(wd / "data_deriv" / "pheno" / "hcp_ya" / "hcp_ya_pheno.csv")
#hcp_ya_pheno
```
